### From common sense, we can say that as our usage increase so does the cost. Hence the monthly bill needs to increase along with GB used per month. So in the next few steps I'll try to find a pattern such that the monthly bill increases along with GB used per month with a little deviation.

```python
min_gbpm = y1 = df2['Total_GB_Used_Per_Month'].min()
max_gbpm = y2 = df2['Total_GB_Used_Per_Month'].max()
min_mnth_bill = x1 = df2['Monthly_Bill'].min()
max_mnth_bill = x2 = df2['Monthly_Bill'].max()
subdf1 = df2[(df2['Total_GB_Used_Per_Month']/df2['Monthly_Bill'] < min_gbpm/min_mnth_bill)]
subdf2 = df2[(df2['Total_GB_Used_Per_Month']/df2['Monthly_Bill'] > max_gbpm/max_mnth_bill)]
subdf3 = df2[(df2['Total_GB_Used_Per_Month']/df2['Monthly_Bill'] > min_gbpm/min_mnth_bill) & (df2['Total_GB_Used_Per_Month']/df2['Monthly_Bill'] < max_gbpm/max_mnth_bill)]
print(subdf1.shape, subdf2.shape, subdf3.shape)
```

```python
m=(y2-y1)/(x2-x1)
c=(x2*y1 - x1*y2)/(x2-x1)
def fnc(x):
    return (m*x + c)
dev = 10
subdf4 = df2[(fnc(df2['Monthly_Bill'])-dev < df2['Total_GB_Used_Per_Month']) & (fnc(df2['Monthly_Bill'])+dev > df2['Total_GB_Used_Per_Month'])]
plt.figure(figsize=(15, 7))
plt.scatter(subdf4['Monthly_Bill'], subdf4['Total_GB_Used_Per_Month'], color='blue', marker='+')
plt.xlabel('Monthly Bill')
plt.ylabel('GB Used per month')
plt.title('Monthly Bill VS GB Used per month')
plt.show()
```

```python
df2 = subdf4.copy()
df2.shape
```

```python
dummies1 = pd.get_dummies(df2.Gender, dtype=int)
dummies1.head()
```

```python
dummies2 = pd.get_dummies(df2.Location, dtype=int)
dummies2.head()
```

```python
df3 = pd.concat([df2, dummies1, dummies2], axis=1)
df3.head()
```

```python
df4 = df3.drop(['Gender', 'Location', 'Total_GB_Used_Per_Month'], axis=1)
df4.head()
```

```python
df5 = df4.drop(['Female', 'Chicago'], axis=1)
df5.head()
```

```python
# df6 = df5.sample(10000)
df6 = df5.copy()
```

```python
X = df6.drop('Churn', axis=1)
y = df6['Churn']
```

```python
X.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify = y)
```

```python
y_train[y_train==0].count()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import SGDClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

models = [
    ('Logistic Regression', LogisticRegression()),
    ('Decision Tree', DecisionTreeClassifier()),
    ('Random Forest', RandomForestClassifier()),
    ('Support Vector Classifier', SVC()),
    ('SGD with Log loss', SGDClassifier())
]

best_model = None
best_mean_score = 0.0

for name, model in models:
    scores = cross_val_score(model, X_train_scaled, y_train, cv=5)
    mean_score = np.mean(scores)
    print(f"{name} - Mean CV score: {mean_score}")
    
    if mean_score > best_mean_score:
        best_mean_score = mean_score
        best_model = model
```

```python
best_model.fit(X_train_scaled, y_train)
best_model.score(X_test_scaled, y_test)
```

```python
X.head(3)
```

```python
def predict_churn(age, sub_months, bill, gb_used, gender, location):
    loc_ind = np.where(X.columns==location)[0][0]

    x = np.zeros(len(X.columns))
    x[0] = age
    x[1] = sub_months
    x[2] = bill
    x[3] = gb_used
    x[4] = gender
    if loc_ind>=0:
        x[loc_ind] = 1

    return best_model.predict([x])[0]
predict_churn(69, 8, 321, 35.61, 1, 'Miami')
```

```python
import xgboost as xgb
from sklearn.metrics import accuracy_score

# Create a DMatrix for training and testing data
dtrain = xgb.DMatrix(X_train, label=y_train)
dtest = xgb.DMatrix(X_test, label=y_test)

# Set the hyperparameters for the XGBoost model
params = {
    'objective': 'reg:logistic',#'binary:hinge',#'binary:logitraw',#'binary:logistic',
    'eval_metric': 'error',
    'max_depth': 3,
    'learning_rate': 0.1
}

# Train the XGBoost model
model = xgb.train(params, dtrain, num_boost_round=10)

# Predict on the test set
y_pred = model.predict(dtest)
y_pred_binary = [1 if p >= 0.5 else 0 for p in y_pred]

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred_binary)
print("Accuracy:", accuracy)
```

```python
import lightgbm as lgb
from sklearn.metrics import accuracy_score

# Create LightGBM Datasets for training and testing data
train_data = lgb.Dataset(X_train, label=y_train)
test_data = lgb.Dataset(X_test, label=y_test, reference=train_data)

# Set the hyperparameters for the LightGBM model
params = {
    'objective': 'binary',
    'metric': 'binary_error',#'binary_logloss',
    'max_depth': 6,
    'learning_rate': 0.1,
    'num_leaves': 30,
    'early_stopping_rounds': 10
}

# Train the LightGBM model
num_round = 100
bst = lgb.train(params, train_data, num_round, valid_sets=[test_data])

# Predict on the test set
y_pred = bst.predict(X_test, num_iteration=bst.best_iteration)
y_pred_binary = [1 if p >= 0.5 else 0 for p in y_pred]

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred_binary)
print("Accuracy:", accuracy)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import SGDClassifier
from sklearn.metrics import accuracy_score

# Standardize the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Create an instance of SGDClassifier for Logistic Regression
clf = SGDClassifier(loss='log_loss', max_iter=1000, random_state=42)

# Train the model using Stochastic Gradient Descent
clf.fit(X_train_scaled, y_train)

# Predict on the test set
y_pred = clf.predict(X_test_scaled)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```
